```python
## read and write on files using pandas
df = pd.read_csv("stock_data.csv")
df
```

```python
 ## now want to skip the row
 df = pd.read_csv('stock_data.csv',skiprows=1)
 ## this will skip the first one row
 df
```

```python
## let say you do not have headers in your data set then 
## and also want to read first n rows
## if there any values that you want to use as a null value then you can replace it
## using na_values parameter this takes array or dictionary as an argument
df = pd.read_csv("stock_data.csv",header=None,names=['Tickers','EPS','REVENUE','PRICE','PEOPLE'],nrows=7,na_values={'EPS':['not available','n.a.'],
    'REVENUE':[-1],'PEOPLE':['not available','n.a.'],'PRICE':['not available','n.a.']})
df
## header = None means there is no header or declaring there is no header
## and names are the columns name or heading of each columns
```

```python
## dropping rows
df = df.drop([0,1])
df
```

```python
df.to_csv('new_stoke_data.csv')
## by default it will take the index also 
# but if you don't want to do so
df.to_csv('new1.csv',index=False)
## you can limit the columns
df.to_csv('new2.csv',index=False,columns=['Tickers','REVENUE'])
## can also remove the header
df.to_csv('new_data.csv',header=False)
```

```python
#read excel
df = pd.read_excel('stock_data.xlsx','Sheet1')
df
```

```python
## we can supply converter argument
def convert_peopleCol(cell):
    if cell=='n.a.':
        return 'sam walton'
    return cell
def convert_epsCol(cell):
    if cell=='not available':
        return None
    return cell
## converters takes dictionary as argument and for each column we can assign functions

df = pd.read_excel('stock_data.xlsx','Sheet1',converters={
    'people':convert_peopleCol,
    'eps':convert_epsCol
})
df
```

```python
## writing to excel file
df.to_excel('new.xlsx',sheet_name='Stocks',startrow=2,startcol=2)
## we can modify the starting  row and col 
```

```python
## writing two dataframes in one excel file into two sheets
df_stocks = pd.DataFrame({
    'tickers': ['GOOGL', 'WMT', 'MSFT'],
    'price': [845, 65, 64 ],
    'pe': [30.37, 14.26, 30.97],
    'eps': [27.82, 4.61, 2.12]
})

df_weather =  pd.DataFrame({
    'day': ['1/1/2017','1/2/2017','1/3/2017'],
    'temperature': [32,35,28],
    'event': ['Rain', 'Sunny', 'Snow']
})
with pd.ExcelWriter('stocks_weather.xlsx') as writer :
    df_stocks.to_excel(writer,sheet_name='stocks')
    df_weather.to_excel(writer,sheet_name='weather')
    
```
